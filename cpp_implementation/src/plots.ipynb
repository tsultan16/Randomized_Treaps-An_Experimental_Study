```python
# read treap sizes and heights from file
treap_sizes = []
treap_heights = []
with open('treap_heights_0.2M.txt', 'r') as f:
    for line in f:
        line = line.replace(',', '')  # remove commas
        treap_sizes.append(int(line.split()[0]))
        treap_heights.append(int(line.split()[1]))
```

```python
import matplotlib.ticker as ticker

# plot treap heights vs L

# log(L) function
def f(x, a=1.0):
    return a*np.log2(x)

# Generate x values
y_values = f(treap_sizes, 2.5)

plt.rcParams.update({'font.size': 14, 'font.family': 'serif'})
plt.plot(treap_sizes, treap_heights, label="Experimental", color='blue', linewidth=0.9)
plt.plot(treap_sizes, y_values, '--', label='$2.5 \log_2 (n_i)$', color='red', linewidth=1.0)

plt.grid(True, which='major', linestyle='-', linewidth=0.1, color='black')
plt.grid(True, which='minor', linestyle=':', linewidth=0.2, color='gray')
plt.minorticks_on()

# Scale x-axis by 10^5
scale_x = 1e5
ticks_x = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_x))
plt.gca().xaxis.set_major_formatter(ticks_x)

plt.xlabel("$n_i$ $(x10^5)$")
plt.ylabel("Height")
plt.legend()
#plt.title("Treap Height vs. Treap Size")
plt.show()
```
